```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# generates RTO frequency regulation signal sent to device fleets (use PJM as example).
# the regulation signal is a 1-dimensional array containing 2-sec signal points.
def get_RegA_signal (ARegA = 1):
    # Arguments:
    # ARegA - 'Assigned Regulation (MW)': assigned hourly regulation quality cleared from the regulation market. 
    # could potentially add variables such as file name, date and hour (not developed yet).
    
    # read Regulation control signal (RegA) from file: 2-sec interval.
    RegA_sig = pd.read_excel('01 2017.xlsx', sheetname='Traditional', 
                             usecols=[0], header=0).head(2030) ### change to 2000 after integration test.###
    RegA_sig = RegA_sig.values # convert from pandas dataframe to numpy array.
    RegA_sig = RegA_sig.flatten()[30:2030] # convert from a 2D numpy array to 1D. ### remove [30,2030] ###

    # When the ARegA is assigned to be 1 as default, it returns normalized [-1, 1] signal. 
    # Otherwise, it returns MW values as signal.
    return ARegA*RegA_sig 
```

```python
# generates response from the device fleet.
# the Current Regulation (CReg) response signal is a 1-dimensional array containing 2-sec signal points.
def get_CReg_response (device_model=0):
    # Arguments:
    # device_model is currently a dummie holder not used in the code.
    # general format for future code: CReg_sig = Device_fleet_model(get_RegA_signal()) .
    
    ### create a dummie response signal below. (80% + 1-min delayed + noise) ###
    
    RegA_sig_delay = pd.read_excel('01 2017.xlsx', sheetname='Traditional', 
                             usecols=[0], header=0).head(2000) 
    RegA_sig_delay = RegA_sig_delay.values # convert from pandas dataframe to numpy array.
    RegA_sig_delay = RegA_sig_delay.flatten() # convert from a 2D numpy array to 1D. 
    dummie_response = RegA_sig_delay * 0.8 + np.random.normal(0,0.1,2000) 
    CReg_sig = dummie_response # it's a dummie for now until the device models are ready for integration.
    
    ### remove above ###
    
    return CReg_sig 
```

```python
# looks up the regulation market clearing price (RMCP).
# the RMCP and its components are stored in a 1-dimensional array. 
def get_RMCP ():
    # Arguments:
    # could potentially add variables such as date and hour (not developed yet).
    
    raw_pr_dataframe = pd.read_excel('historical-ancillary-service-data-2017.xls', 
                                     sheetname='January_2017', header=0).head(10)
#     print(raw_pr_dataframe) 
    
    raw_pr_array = raw_pr_dataframe.values # convert from pandas dataframe to numpy array.
    RMCP = raw_pr_array[0][4] # Regulation Market Clearing Price.
    RMCCP = raw_pr_array[0][6] # Regulation Market Capability Clearing Price. 
    RMPCP = raw_pr_array[0][7] # Regulation Market Performance Clearing Price. 
    
    RM_pr_data = (RMCP, RMPCP, RMCCP) # output as 1-D array.

    return RM_pr_data 
```

```python
# score device fleet's performance (based on PJM Manual 12).
def Reg_perf_score (input_sig = RegA_signal, response_sig = CReg_signal):


    # Arguments:    
    # input_sig - one hour of 2-sec regulation signal.
    # response_sig - one hour of 2-sec response signal sent by device fleet. 
    
    max_corr_array = [] 
    max_index_array = []     

    # Take one hour worth of 2s data slicing at 10s intervals.
    # Need one hour (1800 points) plus 5-min (150) for response signal.
    hr_input_sig = input_sig[0:1950:5]
    hr_response_sig = response_sig[0:1950:5]
    
    
    # In each 5-min of the hour, use max correlation to define "delay", "correlation" & "delay" scores.
    # There are twelve (12) 5-min in each hour.
    for i in range(12): 

        # takes 5-min of the input signal data.
        x = hr_input_sig[30*i:30*(i+1)] 
#         print('x:', x)
        
        # Refresh the array in each 5-min run.
        corr_score = []
    
        # Calculate correlation btw the 5-min input signal and thirty 5-min response signals, 
        # each is delayed at an additional 10s than the previous one; store results.
        # There are thirty (30) 10s in each 5min.
        for j in range(31): 
            y = hr_response_sig[30*i+j:30*(i+1)+j]
            
#             # for debug use            
#             print('y:', y)
#             x_axis_x = np.arange(x.size)
#             plt.plot(x_axis_x, x, "b")
#             plt.plot(x_axis_x, y, "r")
#             plt.show()
            
            # Calculate Pearson Correlation Coefficient btw input and response for the 10s step.
            corr_r = np.corrcoef(x, y)[0,1]
            # Correlation scores stored in an numpy array.
            corr_score = np.append(corr_score, corr_r) 
#         print('corr_score:', corr_score)

        # locate the 10s moment(step) at which correlation score was maximum among the thirty calculated; store it.
        max_index = np.where(corr_score == max(corr_score))[0][0]
        max_index_array = np.append(max_index_array, max_index) # array
        # Find the maximum score in the 5-min; store it.
        max_corr_array = np.append(max_corr_array, corr_score[max_index]) # array
        # Calculate the coincident delay score associated with max correlation in the 5min.
        delay_score = np.absolute((10*max_index_array-5*60)/(5*60)) # array
    
#     # for debug use            
#     print('delay_score:', delay_score)
#     print('max_index_array:', max_index_array)
#     print('max_corr_array:', max_corr_array)
    
    # correlation during periods of zero slope is NOT implemented. Deemed okay without. 
    
    # calculate average "delay" and "correlation" scores for the hour.
    Delay_score = delay_score.mean()
    Corr_score = max_corr_array.mean()

    
    # Calculate Precision Score for the hour (360 x 10s points).
    # Calculate error at 10s steps, store in an array.
    error = np.absolute(hr_response_sig[0:360] - hr_input_sig[0:360])/(np.absolute(hr_input_sig[0:360]).mean())
    
#     # for debug use
#     x_axis_error = np.arange(error.size)
#     plt.scatter(x_axis_error, error)
#     plt.show()

    # Calculate Precision Score as the average error.
    Prec_score = 1 - 1/360*error.sum()
    
    # Calculate the hourly overall Performance Score.
    Perf_score = (Delay_score + Corr_score + Prec_score)/3
    
    
    # Plotting and Printing results        
    x_axis_sig = np.arange(hr_input_sig[0:360].size)
    x_axis_score = np.arange(max_corr_array.size)
    
    plt.figure(1)
    plt.subplot(211)
    plt.plot(x_axis_sig, hr_input_sig[0:360], "b")
    plt.plot(x_axis_sig, hr_response_sig[0:360], "r")
    plt.legend(('RegA signal','CReg response'),loc='lower right')
    plt.show()

    plt.figure(2)
    plt.subplot(212)
    plt.plot(x_axis_score, max_corr_array, "g")
    plt.plot(x_axis_score, delay_score, "m")
    plt.legend(('Correlation score','Delay score'),loc='lower right')
    plt.show()    
     
    print('Delay_score:', Delay_score)
    print('Corr_score:', Corr_score)
    print('Prec_score:', Prec_score)
    print('Perf_score:', Perf_score)
    
    return (Perf_score, Delay_score, Corr_score, Prec_score)
```

```python
# calculate the device fleet's "mileage" for the hour. 
# based on PJM Manual 11. 
# Not used because mileage ratio for traditional regulation is "1".
# would be useful for dynamic regulation OR cost-based regulation offer.
def Reg_mileage (input_sig = RegA_signal):
    
    #Take one hour of RegA data.
    mileage_array = input_sig[1:1800] - input_sig[0:1799]
    RegA_array = input_sig[1:1800]
#     print(mileage_array)
#     x_axis = np.arange(mileage_array.size)
#     plt.plot(x_axis, RegA_array,"b")
#     plt.plot(x_axis, mileage_array,"r")
#     plt.show()
    Mileage = np.absolute(mileage_array).sum()
    
    return Mileage
```

```python
# calculate "Hourly-integrated Regulation MW".
# based on PJM Manual 28 (need to verify definition, not found in manual).
def Hr_int_reg_MW (input_sig = RegA_signal):
    # Arguments:
    # input_sig - one hour of 2-sec regulation signal. 

    #Take one hour of RegA data
    RegA_array = input_sig[1:1800]
#     print(RegA_array)
#     x_axis = np.arange(RegA_array.size)
#     plt.plot(x_axis, RegA_array,"b")
#     plt.show()
    Hourly_Int_Reg_MW = np.absolute(RegA_array).sum()*2/3600
    print(Hourly_Int_Reg_MW)
    
    return Hourly_Int_Reg_MW
```

```python
# calculates hourly "Regulation Market Clearing Price Credit" for a device fleet for the regulation service provided.
# based on PJM Manual 28.
def Reg_clr_pr_credit (RM_pr=get_RMCP(), pf_score=Reg_perf_score()[3], reg_MW=Hr_int_reg_MW()):
    # Arguments:    
    # RM_pr - RMCP price components for the hour.
    # pf_score - performance score for the hour. 
    # reg_MW - "Hourly-integrated Regulation MW" for the hour.
    
    # prepare key parameters for calculation.
    RMCCP = RM_pr[1]
    RMPCP = RM_pr[2]
    
    # "mileage ratio" equals "1" for traditional regulation (i.e. RegA); is >1 for dynamic regulation.
    mi_ratio = 1
    
    print(reg_MW, pf_score, mi_ratio)
    print(RMCCP, RMPCP)
    
    # calculate "Regulation Market Clearing Price Credit" and two components. 
    # minimum perf score is 0.25, otherwise forfeit regulation credit (and lost opportunity) for the hour (m11 3.2.10).
    if pf_score <0.25:
        Reg_RMCCP_Credit = 0
        Reg_RMPCP_Credit = 0
    else:
        Reg_RMCCP_Credit = reg_MW * pf_score * RMCCP
        Reg_RMPCP_Credit = reg_MW * pf_score * mi_ratio * RMPCP 
    Reg_Clr_Pr_Credit = Reg_RMCCP_Credit + Reg_RMPCP_Credit

    print(Reg_RMCCP_Credit, Reg_RMPCP_Credit, Reg_Clr_Pr_Credit)
    
    # "Lost opportunity cost credit" (for energy sources providing regulation) is not considered,
    # because it does not represent economic value of the provided service.
    
    return Reg_Clr_Pr_Credit
```
